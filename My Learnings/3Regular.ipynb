```python
graph = nx.random_regular_graph(2, nodes) #degree = 2
```

```python
plt.figure(figsize=(5, 3))
nx.draw(graph, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')
plt.title("2-Regular Graph")
plt.show()
```

```python
def create_qaoa_circuit(p, n, G):
    
    gamma = ParameterVector('γ',p)
    beta = ParameterVector('β',p)
    
    qc = QuantumCircuit(n, name='q')
    qc.h(range(n)) #superposition
    
    for i in range(p):
        for edge in G.edges(): #problem hamiltonian
            qc.rzz(gamma[i], edge[0], edge[1])
        for qubit in range(n): #mixer Hamiltonian
            qc.rx(2 * beta[i], qubit)
        if i != p-1:
            qc.barrier()
    return qc
```

```python
p=6
```

```python
qaoa = create_qaoa_circuit(p, nodes, graph)
qaoa.measure_all()
qaoa.draw('mpl')
```

```python
from qiskit_ibm_runtime import EstimatorV2

estimator = EstimatorV2(backend=AerSimulator())
```

```python
def obtain_expval(params: list, qaoa: QuantumCircuit, observables: list, estimator) -> float:
    # execute the circuit
    job = estimator.run([(qaoa, observables, params)])
    result = job.result()[0]

    # sum up values
    value = sum(result.data.evs)

    return value
```

```python
from scipy.optimize import minimize

ideal_res = minimize(
    obtain_expval, init_params, args=(qaoa.copy(), observables, estimator), method="cobyla"
)
```

init_params are the initial parameters, we have to find the initial parameters using ML so that <br>
the circuit converges quickly

# USING PENNYLANE OPTIMIZER

```python
from pennylane import qaoa
from pennylane import numpy as numpy
```

```python
import pennylane as qml
```

```python
cost_h, mixer_h = qaoa.maxcut(graph)

print("Cost Hamiltonian", cost_h)
print("Mixer Hamiltonian", mixer_h)
```

```python
def qaoa_layer(gamma, beta):
    qaoa.cost_layer(gamma, cost_h)
    qaoa.mixer_layer(beta, mixer_h)
```

```python
wires = range(4)
depth = 6

def circuit(params, **kwargs):
    for w in wires:
        qml.Hadamard(wires=w)
    qml.layer(qaoa_layer, depth, params[0], params[1])
```

```python
dev = qml.device("qulacs.simulator", wires=wires)

@qml.qnode(dev)
def cost_function(params):
    circuit(params)
    return qml.expval(cost_h)
```

```python
optimizer = qml.GradientDescentOptimizer()
steps = 70
params = numpy.array([[0.5, 0.5, 0.5, 0.5, 0.5, 0.5], [0.5, 0.5, 0.5, 0.5, 0.5, 0.5]], requires_grad=True) 
##cannot initialize with 0, use non-zero values to initialize
```

```python
import time
start = time.time()
for i in range(steps):
    params = optimizer.step(cost_function, params)
end = time.time()
print(f'Time taken : {end-start} seconds')
print("Optimal Parameters")
print(params)
```

```python
@qml.qnode(dev)
def probability_circuit(gamma, beta):
    circuit([gamma, beta])
    return qml.probs(wires=wires)


probs = probability_circuit(params[0], params[1])
```

```python
def create_qaoa_circuit(p, n, G):
    
    gamma = np.array(params[0])
    beta = np.array(params[1])
    
    qc = QuantumCircuit(n, name='q')
    qc.h(range(n)) #superposition
    
    for i in range(p):
        for edge in G.edges(): #problem hamiltonian
            qc.rzz(gamma[i], edge[0], edge[1])
        for qubit in range(n): #mixer Hamiltonian
            qc.rx(2 * beta[i], qubit)
        if i != p-1:
            qc.barrier()
    return qc
```
