# STUDYING CORRELATIONS

```python
import pandas as pd
import csv
import numpy as np
```

```python
trial = pd.read_csv('pennylane_dataset.csv')
trial.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# CORRELATION WITH NODES

```python
plt.figure(figsize=(10, 6))
plt.plot(trial[trial.columns[0]], trial[trial.columns[5]], marker='o', linestyle='-')
plt.xlabel('Nodes')
plt.ylabel('B1(opt)')
plt.title('Plot between Nodes and B1(opt)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(trial[trial.columns[0]], trial[trial.columns[1]], marker='o', linestyle='-')
plt.xlabel('Nodes')
plt.ylabel('Y1(opt)')
plt.title('Plot between Nodes and Y1(opt)')
plt.grid(True)
plt.show()
```

# CORRELATION AMONG PARAMETERS

```python
plt.figure(figsize=(10, 6))
x = trial.iloc[0:1, 1:5]
y = np.array([[1,2,3,4]])
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('Y(opt)')
plt.ylabel('depth')
plt.title('Plot between Y and p (NODES=4) [BETTER CORRELATION WITH DECREASING NODES]')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
x = trial.iloc[4:5, 5:]
y = np.array([[1,2,3,4]])
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('B(opt)')
plt.ylabel('depth')
plt.title('Plot between B and p (NODES=12) [BETTER CORRELATION WITH INCREASING NODES]')
plt.grid(True)
plt.show()
```
